# **Stemming**
### Stemming is a text preprocessing technique in Natural Language Processing (NLP). Specifically, it is the process of reducing inflected form of a word to one so-called “stem,” or root form, also known as a “lemma” in linguistics.

### In other words, Stemming means reducing a word to its root form (called the stem) by chopping off suffixes.

stemming-figure-one-graphic.png




```python
words = [
    "running", "jumps", "easily", "fairly", "studies", "studying", "played",
    "playing", "runner", "happiness", "sadness", "organization", "organized",
    "organizing", "connects", "connected", "connection", "connections",
    "better", "best", "worse", "worst", "fishing", "fished", "fisherman",
    "automation", "automate", "automated", "automation", "argued", "arguing",
    "arguments", "cares", "caring", "careful", "carefully"
]
```

## PorterStemmer

### Key Features & Benefits of Porter Stemmer
1. The algorithm takes off common endings like "-ing," "-ed," and "-ly," changing
"running" to "run" and "happily" to "happi."
2. The stemming process uses several steps to deal with different suffixes, making sure only the right ones are removed.
3. It counts groups of consonants in a word to help decide if certain endings should be taken off.
4. It processes text quickly, which is useful for handling large amounts of data.
5. It provides good results for most common English words and is widely used in NLP projects.
6. By simplifying words to their base forms, it reduces the number of unique words in a dataset, making analysis easier.

### Limitations of Porter Stemmer
1. It can produce stems that are not meaningful, such as turning "iteration" into "iter."
2. The algorithm is primarily designed for English and may not work well with other languages.
3. Compared to other stemmers , it may remove suffixes more aggressively, making words more similar to each other.
4. Different words may be reduced to the same stem, resulting in a loss of meaning.

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
  print(f"{word} -> {stemming.stem(word)}")
```

# **RegexpStemmer Class**

### RegexpStemmer is a rule-based stemmer in NLTK that removes word endings (suffixes) using regular expressions (regex) instead of predefined linguistic rules.

## ⚙️ How it works

### You define a regex pattern that tells the stemmer what to remove from the end of a word.


```python
from nltk.stem import RegexpStemmer
```

```python
stemmer = RegexpStemmer('ing$|ed$|ly$', min=4)
```

```python
stemmer.stem('running')
```

```python
stemmer.stem('happily')
```

# **Snowball Stemmer**

### The Snowball Stemmer is an improved and more advanced version of the Porter Stemmer.
It’s part of the Snowball framework for stemming and supports multiple languages (like English, French, German, Spanish, etc.).

```python
from nltk.stem import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
stemmer.stem('running')
```

```python
stemmer.stem('happily')
```

```python
for word in words:
  print(f"{word} -> {stemmer.stem(word)}")
```

```python
stemmer.stem("fairly")
```

```python
stemmer.stem("sportingly")
```
